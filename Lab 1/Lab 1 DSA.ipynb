```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

df = pd.read_csv("education_career_success.csv")
```

```python
bins = [0, 2.5, 3.0, 3.5, 4.0]
labels = ['Low', 'Average', 'Good', 'Excellent']
df['GPA_Category'] = pd.cut(df['University_GPA'], bins=bins, labels=labels)
df['GPA_Category']
```

```python
df.groupby('Internships_Completed')['Job_Offers'].mean()
```

```python
df.groupby('Internships_Completed')['Starting_Salary'].mean()
```

```python
df['Starting_Salary'].describe()
```

```python
df[['University_GPA', 'Starting_Salary']].corr()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Starting_Salary"], bins=30, kde=True, color="blue")
plt.title("Distribution of the starting salary")
plt.xlabel("Salary ($)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
df.groupby('GPA_Category', observed=True)['Starting_Salary'].mean()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[["Starting_Salary"]])
plt.title("Outlier Analysis for Key Variables")
plt.xticks(rotation=15)
plt.show()
```

```python
correlation_matrix = df.select_dtypes(include=["number"]).corr()
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
df.groupby('GPA_Category', observed=True)['Job_Offers'].mean()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["University_GPA"], y=df["Job_Offers"], alpha=0.5, color="blue", label="GPA vs Job Offers")
sns.scatterplot(x=df["Internships_Completed"], y=df["Job_Offers"], alpha=0.5, color="red", label="Internships vs Job Offers")
plt.title("Impact of GPA and Internships on Job Offers")
plt.xlabel("GPA / Internships")
plt.ylabel("Number of Job Offers")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df.groupby('University_Ranking')['Starting_Salary'].mean()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df["University_GPA"], y=df["Starting_Salary"], alpha=0.5, color="blue", label="GPA vs Salary")
sns.scatterplot(x=df["University_Ranking"], y=df["Starting_Salary"], alpha=0.5, color="red", label="Ranking vs Salary")
plt.title("Impact of University Ranking and GPA on Starting Salary")
plt.xlabel("GPA / University Ranking")
plt.ylabel("Starting Salary ($)")
plt.legend()
plt.grid(True)
plt.show()
```
