```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Income_expense_data.csv')
```

```python
print(df.info())
```

```python
df
```

```python
df_cleaned = df.sample(n=10000, random_state=42).reset_index(drop=True)
```

```python
print("Number of duplicates:", df_cleaned.duplicated().sum())
```

```python
print("Missing values before cleaning:\n", df_cleaned.isnull().sum())
```

```python
df_cleaned1 = df_cleaned.dropna(subset=['Expense', 'Income'])
```

```python
print(df_cleaned1.info())
```

```python
print("Missing values after cleaning:\n", df_cleaned1.isnull().sum())
```

```python
print(df_cleaned1.dtypes)
```

```python
print(df_cleaned1.info())
```

```python
df_cleaned1.loc[:, 'Expense'] = df_cleaned1['Expense'].fillna(df_cleaned1['Expense'].mean())
df_cleaned1.loc[:, 'Income'] = df_cleaned1['Income'].fillna(df_cleaned1['Income'].median())

df_cleaned1
```

```python
print("Missing values after cleaning:\n", df_cleaned1.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Household expenditures, summary-level categories', y='Expense', data=df_cleaned1)
plt.xticks(rotation=90)
plt.show()
```

```python
Q1 = df_cleaned1['Expense'].quantile(0.25)
Q3 = df_cleaned1['Expense'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(Q1)
print(Q3)
print(IQR)
print(lower_bound)
print(upper_bound)
```

```python
df_cleaned1 = df_cleaned1[(df_cleaned1['Expense'] >= lower_bound) & (df_cleaned1['Expense'] <= upper_bound)]

df_cleaned1
```

```python
df_cleaned1['Expense_scaled'] = (df_cleaned1['Expense'] - df_cleaned1['Expense'].min()) / (df_cleaned1['Expense'].max() - df_cleaned1['Expense'].min())
```

```python
df_cleaned1['Family income'] = df_cleaned1['Family income'].str.strip()
```

```python
print(df_cleaned1['Family income'].unique())
```

```python
df_cleaned2 = df_cleaned1.copy()

# Заменяем значения в 'Family income'
income_map = {
    'Under $10,000': 1,
    '$10,000 and over': 2,
    '$15,000 and over': 3,
    '$20,000 and over': 4,
    '$25,000 and over': 5,
    '$30,000 and over': 6,
    'Median total income': 7,
    '$70,000 and over': 8,
    '$75,000 and over': 9,
    '$80,000 and over': 10,
    '$90,000 and over': 11,
    'All income groups': 0
}

df_cleaned2['Family income'] = df_cleaned2['Family income'].map(income_map)

df_cleaned2['Family income'] = df_cleaned2['Family income'].fillna(0)

print(df_cleaned2['Family income'].unique())
```

```python
df_cleaned2
```

```python
print(df_cleaned2.describe())
```

```python
correlation = df_cleaned2[['Expense', 'Income']].corr()
print("Correlation between Expense and Income:\n", correlation)
```
