```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Walmart_customer_purchases.csv")
```

```python
df_sample = df.sample(n=5000, random_state=42)
```

```python
df_sample.shape
```

```python
df_sample["Purchase_Date"] = pd.to_datetime(df_sample["Purchase_Date"])
```

```python
df_sample.describe()
```

```python
df_corr = df_sample.select_dtypes(include=['number'])
plt.figure(figsize=(10,6))
sns.heatmap(df_corr.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
ltv = df_sample.groupby("Customer_ID")["Purchase_Amount"].sum().reset_index()
ltv.columns = ["Customer_ID", "Lifetime_Value"]
print("Top 5 LTV customers:")
ltv.head()
```

```python
retention_rate = df_sample["Repeat_Customer"].value_counts(normalize=True).get("Yes", 0) * 100
print(f"Customer Retention Rate: {retention_rate:.2f}%")
```

```python
churn_rate = 100 - retention_rate
print(f"Customer Churn Rate: {churn_rate:.2f}%")
```

```python
plt.figure(figsize=(6,6))
plt.pie([retention_rate, churn_rate], labels=["Retained", "Churned"], autopct="%1.1f%%", colors=["lightblue", "salmon"])
plt.title("Customer Retention vs Churn Rate")
plt.show()
```

```python
average_order_value = df_sample["Purchase_Amount"].mean()
print(f"Average Order Value: {average_order_value:.2f}")
```

```python
total_revenue = df_sample["Purchase_Amount"].sum()
total_revenue
```

```python
purchase_frequency = df_sample["Customer_ID"].value_counts().mean()
print(f"Purchase Frequency per Customer: {purchase_frequency:.2f}")
```

```python
average_rating = df_sample["Rating"].mean()
print(f"Average Product Rating: {average_rating:.2f}")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_sample["Purchase_Amount"], bins=30, kde=True)
plt.xlabel("Purchase Amount")
plt.ylabel("Count")
plt.title("Distribution of Purchase Amounts")
plt.show()
```

```python
df_sample["Month"] = df_sample["Purchase_Date"].dt.to_period("M")
monthly_sales = df_sample.groupby("Month")["Purchase_Amount"].sum()
monthly_sales
```

```python
plt.figure(figsize=(12, 5))
monthly_sales.plot(marker="o", linestyle="-")
plt.xlabel("Month")
plt.ylabel("Total Purchase Amount")
plt.title("Monthly Purchase Trends")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x="Gender", y="Purchase_Amount", data=df_sample)
plt.title("Purchases by Gender")
plt.xlabel("Gender")
plt.ylabel("Purchase Amount")
plt.show()
```

```python
plt.figure(figsize=(15, 4))
sns.boxplot(x="Age", y="Purchase_Amount", data=df_sample)
plt.title("Purchases by Age Group")
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.show()
```

```python
category_avg_purchase = df_sample.groupby("Category")["Purchase_Amount"].mean()
category_avg_purchase
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x="Category", y="Purchase_Amount", data=df_sample)
plt.title("Purchases by Product Category")
plt.xlabel("Category")
plt.ylabel("Purchase Amount")
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df["Repeat_Customer"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, cmap="Pastel1")
plt.title("Repeat Customer Distribution")
plt.ylabel("")
plt.show()
```
